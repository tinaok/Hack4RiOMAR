```python
import matplotlib.pyplot as plt
import numpy as np
import healpy as hp

# Base HEALPix resolution
nside_base = 1  # corresponds to 12 pixels
npix_base = hp.nside2npix(nside_base)

# Higher resolution for zoom-in
nside_zoom = 4  # each base pixel subdivides into 16
zoom_factor = (nside_zoom // nside_base) ** 2

# Choose pixel 5 to zoom into
base_pixel = 5
start = base_pixel * zoom_factor
subpixels = np.arange(start, start + zoom_factor)

# Plotting setup
fig = plt.figure(figsize=(12, 6))

# Plot base level HEALPix grid
ax1 = fig.add_subplot(1, 2, 1, projection='mollweide')
hp.mollview(np.arange(npix_base), nest=True, title="Base Level (Nside=1)", hold=True, fig=1, cbar=False)
vec_base = hp.pix2vec(nside_base, np.arange(npix_base), nest=True)
theta_base, phi_base = hp.vec2ang(vec_base)
hp.projtext(phi_base - np.pi, 0.5 * np.pi - theta_base,
            [str(i) for i in range(npix_base)],
            lonlat=False, fontsize=10, color='black')

# Plot nested structure of pixel 5
ax2 = fig.add_subplot(1, 2, 2, projection='mollweide')
hp.mollview(np.zeros(hp.nside2npix(nside_zoom)), nest=True,
            title="Nested Indexing of Pixel 5 (Nside=4)", hold=True, fig=1, cbar=False)
vec_sub = hp.pix2vec(nside_zoom, subpixels, nest=True)
theta_sub, phi_sub = hp.vec2ang(vec_sub)
hp.projtext(phi_sub - np.pi, 0.5 * np.pi - theta_sub,
            [f"5.{i}" for i in range(zoom_factor)],
            lonlat=False, fontsize=9, color='darkgreen')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import healpy as hp

# Base HEALPix resolution
nside_base = 1  # corresponds to 12 pixels
npix_base = hp.nside2npix(nside_base)

# Higher resolution for zoom-in
nside_zoom = 4  # each base pixel subdivides into 16
zoom_factor = (nside_zoom // nside_base) ** 2

# Choose pixel 5 to zoom into
base_pixel = 5
start = base_pixel * zoom_factor
subpixels = np.arange(start, start + zoom_factor)

# Plotting setup
fig = plt.figure(figsize=(12, 6))

# Plot base level HEALPix grid
hp.mollview(np.arange(npix_base), nest=True, title="Base Level (Nside=1)", hold=True, fig=1, cbar=False, sub=(1, 2, 1))
vec_base = np.stack(hp.pix2vec(nside_base, np.arange(npix_base), nest=True), axis=-1)
theta_base, phi_base = hp.vec2ang(vec_base)

# Add text labels for base pixels
for i in range(npix_base):
    x = phi_base[i] - np.pi
    y = 0.5 * np.pi - theta_base[i]
    hp.projtext(x, y, str(i), lonlat=False, fontsize=10, color='black')

# Plot nested structure of pixel 5
hp.mollview(np.zeros(hp.nside2npix(nside_zoom)), nest=True,
            title="Nested Indexing of Pixel 5 (Nside=4)", hold=True, fig=1, cbar=False, sub=(1, 2, 2))
vec_sub = np.stack(hp.pix2vec(nside_zoom, subpixels, nest=True), axis=-1)
theta_sub, phi_sub = hp.vec2ang(vec_sub)

# Add text labels for subpixels
for i, (theta, phi) in enumerate(zip(theta_sub, phi_sub)):
    x = phi - np.pi
    y = 0.5 * np.pi - theta
    hp.projtext(x, y, f"5.{i}", lonlat=False, fontsize=9, color='darkgreen')

plt.tight_layout()
plt.show()
```
